```python
#POLYNOMIAL REGRESSION
import random

#from matrix-2.txt
x1 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
y1 = [0, 0.8, 0.9, 0.2, -0.8, -1, -0.8, -0.4, -0.2]

x1 = np.array(x1)
y1 = np.array(y1)

plt.figure(figsize=(12, 8))

#looping
for i in range(1, 7):
    #count coefficient of the polynomial for the regression
    y_est = np.polyfit(x1, y1, i)
    
    #generate double plot
    plt.subplot(2, 3, i)
    
    plt.plot(x1, y1, 'o')
    
    plt.plot(x1, np.polyval(y_est, x1))
    
    plt.title(f"Polynomial Order {i}")
    
plt.tight_layout()
plt.show()    
```

```python
#CUBLIC SPLINE INTERPOLATION
from scipy.interpolate import CubicSpline

x2 = [0, 1, 2]
y2 = [1, 3, 2]

#make one cublic spline function based on x and y values
#we need natural boundary condition (bc_type="natural")
f = CubicSpline(x2, y2, bc_type="natural")

#make 100 dots from range x = 0 till x = 2
x_new = np.linspace(0, 2, 100)
print(x_new)

#count new y value using function f as stated above
y_new = f(x_new)

plt.plot(x_new, y_new, 'b')
plt.plot(x2, y2, 'ro')

plt.title("Cubic Spline Interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
