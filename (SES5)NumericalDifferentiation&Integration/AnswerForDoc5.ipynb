```python
# PROBLEM 1 (RIEMANN INTEGRAL)
import numpy as np

#example: 4x^4 + 2x^3 - 6x
def f(x):
    return 4*x**4 + 2*x**3 - 6*x

a = -3
b = 3
n = 30
totalWidth = (b - a) / (n - 1)
x = np.linspace(a, b, n)
y = f(x)

xMid = (x[:n - 1] + x[1: ]) / 2
yMid = f(xMid)

leftRiemann = totalWidth * sum(y[:n - 1])
rightRiemann = totalWidth * sum(y[1: ])
midPoint = totalWidth * sum(yMid)

print(f"Left Riemann: {leftRiemann}")
print(f"Right Riemann = {rightRiemann}")
print(f"Midpoint: {midPoint}")
```

```python
# PROBLEM 2 (TRAPEZOID RULE)
import numpy as np

def f(x):
    return np.sin(x)**2 - np.sin(x) + 2

a = -np.pi
b = 2 * np.pi
n = 27
totalWidth = (b - a) / (n - 1)
x = np.linspace(a, b, n)
y = f(x)

trapezoid = totalWidth / 2 * (y[0] + 2 * sum(y[1 : n - 1]) + y[n - 1])
print(f"Result: {trapezoid}")
```
