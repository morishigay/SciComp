# Riemann Integral

Riemann adalah sebuah cara untuk menghitung luas daerah di bawah sebuah kurva.

**Konsep utama:** Membagi area menjadi $n$ jumlah kotak, lalu menghitung luas masing-masing kotak.

Ada 4 jenis yang akan dipelajari:

1. Left Riemann
2. Right Riemann
3. Midpoint Rule
4. Trapezoid Rule

### Logic

1. Tentukan function dan range luas.
2. Tentukan jumlah kotak yang ingin digunakan.
3. Hitung lebar masing-masing kotak:

$$\Delta x = \frac{b - a}{n}$$



4. Hitung panjang masing-masing kotak.
5. Gunakan rule yang diinginkan.

```python
import numpy as np

#example
def f(x):
    return x**2 - 5

#to find the area we need x0 (a) and x1 (b)
a = 2
b = 5

#store box values in form of grid points
n = 5

#find the width of the box
totalWidth = (b - a) / (n - 1)

#find the distance from the x-axis
x = np.linspace(a, b, n)
y = f(x)

#for midpoint
xMid = (x[:n - 1] + x[1: ]) / 2
yMid = f(xMid)

leftRiemann = totalWidth * sum(y[:n - 1]) #explain why
rightRiemann = totalWidth * sum(y[1: ]) #explain why 2
midPoint = totalWidth * sum(yMid)
trapezoid = totalWidth / 2 * (y[0] + 2 * sum(y[1:n - 1]) + y[n - 1])

print(f"Left Riemann: {leftRiemann}")
print(f"Right Riemann: {rightRiemann}")
print(f"Midpoint: {midPoint}")
print(f"Trapezoid: {trapezoid}")
```
