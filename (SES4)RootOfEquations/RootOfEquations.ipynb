```python
# BISEXUAL METHOD
import numpy as np

#example
def f(x):
    return x**2 - 4

def bisection(a, b, tol):
    # check if f(a) is is different from f(b) 
    if (np.sign(f(a)) == np.sign(f(b))):
        print("Interval does not bracket a root")
        return
    
    #mid point formula
    c = (a + b) / 2
    
    #check exact root of c
    if (f(c) == 0): # or np.abs(f(c)) < tol):
        print("Root = ", c) #root found
        return
    
    #check tolerance
    if (np.abs(f(c)) < tol):
        print("Root = ", c)
        return
    
    #check which to replace with c
    if (np.sign(f(a)) == np.sign(f(c))):
        bisection(c, b, tol)
    elif (np.sign(f(b)) == np.sign(f(c))): 
        bisection(a, c, tol)

bisection(0, 5, 0.01)
```

```python
# NEWTON-RAPHSON METHOD
import numpy as np

#examples
def f(x):
    return x**2 - 4

def g(x):
    return 2*x

def newtonRhapson(x0, tol, maxIter=100):
    for i in range(maxIter):
        #edge case
        if (g(x0) == 0):
            print("Derivative is 0, cannot divide by zero")
            return
        
        #formula
        x1 = x0 - f(x0) / g(x0)
        
        #for showing off iteration process
        print(f"Iteration {i + 1}: {x1}")
        
        #check tolerance
        if (np.abs(f(x1)) < tol):
            print("Root = ", x1)
            return
        x0 = x1
        print("Did not converge")
        
newtonRhapson(3, 0.01)
```
